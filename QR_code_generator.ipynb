# **Py QR code generator**

Easily generate a QR code for anything you would like.

```python
# This notebook demonstrates how to generate a QR code using Python.
```

```python
#Install 'qrcode' library and import `qrcode` and `PIL.Image` modules

!pip install qrcode
import qrcode
from PIL import Image
```

```python
#Enter input for the data to be encoded, create a QR code object, adds the data, and generate the QR code image.

data = input("Enter what you want to create QR code for : ")
qr = qrcode.QRCode(version=3, box_size=8, border=2)
qr.add_data(data)
qr.make(fit=True)
image = qr.make_image(fill="Black", back_color="GreenYellow")
```

```python
#Saves the generated QR code image to a file named 'qr_code.png'.

image.save("qr_code.png")
```
